```python
import numpy as np
import cv2
import os 
```

```python
def Read_Data(train_path, test_path):
    train_data = np.asarray([i.split() for i in open(train_path).readlines()], dtype='float')
    test_data  = np.asarray([i.split() for i in open(test_path).readlines()], dtype='float')
  
    train_X, train_y = train_data[:, :-1], train_data[:, -1:]
    test_X, test_y = test_data[:, :-1], test_data[:, -1:]
    
    return train_X, train_y, test_X, test_y
```

```python
def sign(x):
    return (x>=0) and 1 or -1

def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
def sdc_gradient(x, y, w):
    grad = sigmoid(-1*np.dot(w, y*x))*-1*y*x
    
    return grad
```

```python
def gradient(x, y, w):
    
    error = 0
    for i, data in enumerate(x):
        error += sigmoid(np.dot(w, -1*y[i]*data))*-1*y[i]*data
    
    return error/len(x)
```

```python
def Logistic_Regression(X, y, w_init, learning_rate, iterations):
    size = X.shape[0]
    w = w_init
    
    j = 0
    for i in range(iterations):
        grad = gradient(X, y, w)
        print(grad)
        w = w - learning_rate*grad
        
        j = j%size
    return w
```
